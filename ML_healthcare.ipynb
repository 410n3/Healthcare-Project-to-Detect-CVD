## Build a baseline model to predict the risk of a heart attack using a logistic regression and random forest and explore the results while using correlation analysis and logistic regression (leveraging standard error and p-values from statsmodels) for feature selection


```python
hc_cato.columns
```

```python
catos=['sex', 'fbs', 'restecg', 'cp', 'exang', 'slope', 'ca', 'thal']
```

```python
hc_final=pd.get_dummies(hc,columns=catos,drop_first=True)
hc_final.head()
```

```python
hc_final.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = hc_final.drop('target',axis=1)
Y= hc_final['target']
```

```python
X_train, X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
print(" Training data shape: ",X_train.shape,"\n","Test data shape: ",X_test.shape,"\n","Training target shape: ",y_train.shape,"\n","Test target shape: ",y_test.shape)
```

#### we have splited the data into train data and test data

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(fit_intercept=False,penalty='none',solver="newton-cg")
```

```python
lr_model=lr.fit(X_train,y_train)
```

#### the accuracy score of  Logistic Regression is 88% which is good so we have to compair with test data whether the model is over fitting or not

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report,plot_confusion_matrix
```

```python
lr_model_pred=lr_model.predict(X_train)
accuracy_lr=accuracy_score(y_train,lr_model_pred)
print("accuracy of the model is : ",accuracy_lr)
```

#### here we can see the accuracy of the model is 88% which is good so ww have to check how well the model is predicting with confusion matrix

```python
sns.heatmap(confusion_matrix(y_train,lr_model_pred),annot=True,fmt="d")
plt.xlabel("actual")
plt.ylabel("predicted")
```

```python
print(classification_report(y_train,lr_model_pred))
```

### now checking with the test data whether the model is overfitting or not

```python
lr_model_pred_test=lr_model.predict(X_test)
accuracy_lr_test=accuracy_score(y_test,lr_model_pred_test)
print("accuracy of the model is : ",accuracy_lr_test)
```

```python
sns.heatmap(confusion_matrix(y_test,lr_model_pred_test),annot=True,fmt="d")
plt.xlabel("actual")
plt.ylabel("predicted")
```

### as from this we can see that there is not much overfitting is there so its a good model

```python
pd.DataFrame(zip(X_train.columns, np.transpose(lr_model.coef_)), columns=['features', 'coef']) 
```

#### now we have to choose some significant features that can help us with prediction

```python
import statsmodels.api as sm
logit_m=sm.Logit(y_train,X_train)
results=logit_m.fit(method='newton')
```

```python
print(results.summary2())
```

### here we can see some features are insignificant which is grater than 0.05 so we use rfe to choose 8 significant varaibles

```python
from sklearn.feature_selection import RFE
```

```python
logreg = LogisticRegression()
rfe=RFE(estimator=logreg,n_features_to_select=8,verbose=True)
rfe= rfe.fit(X_train,y_train)
```

```python
pd.set_option('display.max_rows',500)
best_fe_rfe=pd.DataFrame({'features :':list(X_train.columns),
                                     'Support':rfe.support_,
                                     'Ranking':rfe.ranking_})
best_fe_rfe.sort_values(by='Ranking')
```

```python
X_train_rfe1=X_train[["cp_2","ca_3","ca_2","ca_1","slope_1","exang_1","cp_1",'thal_2']]
X_test_rfe1=X_test[["cp_2","ca_3","ca_2","ca_1","slope_1","exang_1","cp_1",'thal_2']]
```

#### now using these feature we again buil Logistic model

```python
lr_model_rfe=lr.fit(X_train_rfe1, y_train)
y_pred_rfe = lr_model_rfe.predict(X_train_rfe1)
accuracy_rfe = accuracy_score(y_train, y_pred_rfe)
print("accuracy train data :",accuracy_rfe)
y_pred_test_rfe = lr_model_rfe.predict(X_test_rfe1)
accuracy_rfe1 = accuracy_score(y_test, y_pred_test_rfe)
print("accuracy test data :",accuracy_rfe1)

```

### here we can see the accuracy of the model is  87% which is poor now we have to use Random Forrest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf.fit(X_train, y_train)
rf_pred=rf.predict(X_test)
print(classification_report(y_test,rf_pred))
```

### here random forrest is giving us 77 % of accuracy  and Logistic regression is giving us 87% accuracy

```python
#testingall the models with test data
lr.fit(X_train,y_train)
lr_t_pred=lr.predict(X_test)
print("For Logestic Regression",classification_report(y_test,lr_t_pred))
lr.fit(X_train_rfe1,y_train)
lr_t2_pred=lr.predict(X_test_rfe1)
print("For Logestic Regression with rfe",classification_report(y_test,lr_t2_pred))
rf.fit(X_train_rfe1,y_train)
rf_t_pred=rf.predict(X_test_rfe1)
print("For random forest with rfe",classification_report(y_test,rf_t_pred))
rf.fit(X_train,y_train)
rf_t2_pred=rf.predict(X_test)
print("For random forest without rfe",classification_report(y_test,rf_t2_pred))


```
